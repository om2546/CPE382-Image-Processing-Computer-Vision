```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Intensity Manipulation

### 1.1 Bright Image

```python
bright_img = cv2.imread("brightImg.png") # Read in the image
bright_img = cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB) # Convert to RGB
plt.imshow(bright_img) # Display the bright image
plt.show()
```

```python
def plot_hist_rgb(img):
    for i in range(img.shape[2]):
        plt.subplot(3,1,i+1)
        plt.hist(img[:,:,i].ravel(),256,[0,256])
        plt.title("Channel {}".format(i))
    plt.tight_layout()
    plt.show()
```

```python
plot_hist_rgb(bright_img) # Plot a histogram of the RGB channels r = 0, g = 1, b = 2
```

Using exponential operator

```python
def exponential(r,c,e):
    return np.round(c*e**r)
```

Adjust the `exponential` value 

```python
e = 1.02
c = 255/(e**255)
```

```python
test_range = np.arange(0, 256) # Create a test range of values from 0 to 255
test_range  = np.uint8(exponential(test_range, c,e)) # Apply exponential function
plt.title('Exponential Function')
plt.plot(test_range) # Plot the function
plt.show()
```

Apply `exponential` function

```python
bright_img_exp = np.uint8(exponential(bright_img, c, e)) # Apply exponential function to bright image

plt.subplot(1,2,1)
plt.imshow(bright_img) # Display original image
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(bright_img_exp) # Display transformed image
plt.title('Exponential Transformed Image')

plt.tight_layout
plt.show()
```

```python
plot_hist_rgb(bright_img_exp) # Plot histogram of bright image after applying exponential function
```

### 1.2 Dark Image

```python
dark_img = cv2.imread("darkImg.png") # Read in the image
dark_img = cv2.cvtColor(dark_img, cv2.COLOR_BGR2RGB) # Convert to RGB
plt.imshow(dark_img) # Display the dark image
```

```python
plot_hist_rgb(dark_img) # Plot a histogram of the RGB channels r = 0, g = 1, b = 2
```

Using logarithm operator

```python
def logarithmic(r,e,c):
    return np.round(np.emath.logn(e,r+1)*c)
```

Adjust `base` value of log function

```python
e = np.e
c = 255/np.emath.logn(e,255)
```

```python
logarithmic(0,e,c) , logarithmic(255,e,c)
```

```python
test_range = np.arange(0, 256) # Create a test range of values from 0 to 255
test_range  = np.uint8(logarithmic(test_range, e, c)) # Apply logarithmic function
plt.title('Logarithmic Function')
plt.plot(test_range) # Plot the function
plt.show()
```

Apply  `logarithmic` function

```python
dark_img_log = np.uint8(logarithmic(dark_img, e, c)) # Apply logarithmic function to dark image

plt.subplot(1,2,1)
plt.imshow(dark_img) # Display original image
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(dark_img_log) # Display transformed image
plt.title('Logarithmic Transformed Image')

plt.tight_layout()
plt.show()
```

```python
plot_hist_rgb(dark_img_log) # Plot histogram of dark image after applying logarithmic function
```
