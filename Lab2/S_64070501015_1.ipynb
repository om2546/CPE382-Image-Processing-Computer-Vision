```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Intensity Manipulation

### 1.1 Bright Image

```python
bright_img = cv2.imread("brightImg.png") # Read in the image
bright_img = cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB) # Convert to RGB
plt.imshow(bright_img) # Display the bright image
```

```python
def plot_hist_rgb(img):
    for i in range(img.shape[2]):
        plt.subplot(3,1,i+1)
        plt.hist(img[:,:,i].ravel(),256,[0,256])
        plt.title("Channel {}".format(i))
    plt.tight_layout()
    plt.show()
```

```python
plot_hist_rgb(bright_img) # Plot a histogram of the RGB channels r = 0, g = 1, b = 2
```

Using exponential operator

```python
def exponential(r,c,e):
    return np.round(c*e**r)
```

Adjust the `exponential` value 

```python
e = 1.02
c = 255/(e**255)
```

```python
test_range = np.arange(0, 256) # Create a test range of values from 0 to 255
test_range  = np.uint8(exponential(test_range, c,e)) # Apply exponential function
plt.title('Exponential Function')
plt.plot(test_range) # Plot the function
```

Apply `exponential` function

```python
bright_img_exp = np.uint8(exponential(bright_img, c, e)) # Apply exponential function to bright image

plt.subplot(1,2,1)
plt.imshow(bright_img) # Display original image
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(bright_img_exp) # Display transformed image
plt.title('Exponential Transformed Image')
plt.tight_layout
plt.show()
```

```python
plot_hist_rgb(bright_img_exp) # Plot histogram of bright image after applying exponential function
```
